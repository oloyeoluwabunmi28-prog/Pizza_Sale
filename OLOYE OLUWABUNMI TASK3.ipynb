```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
order_detail = pd.read_csv("/content/order_details.csv")

order_detail
```

```python
orders = pd.read_csv("/content/orders.csv")

orders
```

```python
pizza_type = pd.read_csv("/content/pizza_types.csv", encoding="latin1")


pizza_type
```

```python
pizzas = pd.read_csv("/content/pizzas.csv")


pizzas
```

```python
dic = pd.read_csv("/content/data_dictionary.csv")

dic
```

#Merging All the data together

```python
#Merging order details and order id together.

merge_order= pd.merge(order_detail, orders, on='order_id', how='outer')


merge_order
```

```python
#Merging pizzatype and pizza type id together

merge_pizza= pd.merge(pizza_type,pizzas, on='pizza_type_id', how='outer')


merge_pizza
```

```python
#Joining the two merged above together

merged_data = pd.merge(merge_order,merge_pizza, on='pizza_id', how='inner')


merged_data
```

```python
merged_data.head()
```

```python
#step1 The total revenue/sales


merged_data["Total sales"] = merged_data["quantity"] * merged_data["price"]
```

```python
merged_data.head()
```

```python
#Step 2. To find the total quantity sold



total_quantity_sold = merged_data["quantity"].sum()


total_quantity_sold
```

```python
#Step 3. To find the total orders



total_orders = len(merged_data) #counts the number of orders

total_orders
```

```python
total_orders =merged_data["order_id"].nunique()  #returns an array of orders id

total_orders
```

```python
#Step4. To find pizza types they sell



pizza_sold_by_type = merged_data.groupby("name")["price"].sum().reset_index()



pizza_sold_by_type
```

```python
#Step 5. Find the average price of the pizzas



avg_price_by_pizza = merged_data.groupby("name")["price"].mean().reset_index()


avg_price_by_pizza
```

```python
#Step 6 To find the peak hours of sales


merged_data["hour"] = pd.to_datetime(merged_data["time"]).dt.hour
```

```python
sales_by_peak_hour = merged_data.groupby("hour")["quantity"].sum().reset_index()


sales_by_peak_hour
```

```python
#Step 7a. To Find the Total sales made on each day of the week


merged_data["date"] = pd.to_datetime(merged_data["date"])

merged_data["days_of_the_week"] = merged_data["date"].dt.day_name()

total_sales_by_day = merged_data.groupby("days_of_the_week")["price"].sum().reset_index()


total_sales_by_day
```

```python
#Visualizing the results

plt.bar(total_sales_by_day["days_of_the_week"], total_sales_by_day["price"])
plt.xticks(rotation=90)
plt.title("Weekly Sales report")
plt.xlabel ("Days of the Week")
plt.ylabel("Total price")


plt.show()
```

```python
#Step 7b. To find the day of the week when sales are made most

best_day =total_sales_by_day.loc[total_sales_by_day["price"].idxmax()]

best_day
```

```python
#Step 8. Find the top 5 bestselling pizzas


top_best = merged_data.groupby("name")["quantity"].sum().reset_index()


top_best =top_best.sort_values("quantity", ascending=False)


top_best
```

```python
# the top 5 will be


top_best.head()
```

```python
#Step 9 To find the sales made in each month

merged_data["date"]=pd.to_datetime(merged_data["date"])


merged_data["month"]=merged_data["date"].dt.month_name()


sales_by_month = merged_data.groupby("month")["price"].sum().reset_index()


sales_by_month
```

```python
#another method adding the year


merged_data["month"] =merged_data["date"].dt.month_name()

merged_data["year"] = merged_data["date"].dt.year

sales_by_month_year =merged_data.groupby(["year","month"])["price"].sum().reset_index()


sales_by_month_year
```

```python
plt.plot(sales_by_month["month"], sales_by_month["price"],color="red",marker="o")
plt.title("Pizza Monthly Sales")
plt.xticks(rotation=90)
plt.xlabel("Month of the year")
plt.ylabel("Pizza Price")

plt.show()
```

#The plot shows a clear peak of ales during the summer months,especially July  and a slow Sales in October.

#This suggest that  pizza Sales are higher during summer month(July) when people are likely to order food for outdoor gathering and in contrast, slow Sales in September and October when consume are being cautious with their spending before the holiday season/Christmas.

```python
#Step10. To know if there Pizza type with low sales on the menu

pizza_type_low_sales= merged_data.groupby("name")["quantity"].sum().reset_index()

pizza_type_low_sales =pizza_type_low_sales.sort_values("quantity",ascending =True)

pizza_type_low_sales
```

```python
pizza_type_low_sales.head(5)
```

```python
#To know pizza that is not doing well/low sale by revenue using pizza type id.

merged_data ["revenue"] = merged_data["quantity"] * merged_data["price"]

pizza_type_revenue = merged_data.groupby("pizza_type_id")["revenue"].sum().reset_index()

pizza_type_revenue =pizza_type_revenue.sort_values("revenue",ascending =True)


low_revenue_pizza = pizza_type_revenue.head(5)

low_revenue_pizza
```
